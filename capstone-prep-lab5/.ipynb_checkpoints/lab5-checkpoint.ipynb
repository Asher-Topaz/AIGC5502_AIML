```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("wine.data",header=None)
df.columns = ['Label', 'Alcohol', 'Malic Acid', 'Ash','Alcalinity of ash', 'Magnesium', 
              'Total phenols', 'Flavanoids' , 'Nonflavanoid phenols', 'Proanthocyanins',
                'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
df.head(10)
```

```python
df.plot.scatter(x= 'Alcohol', y = 'Label', title = 'Label vs Alcohol')

df.plot.scatter(x = 'Malic Acid', y = 'Label', title = 'Label vs Malic')

df.plot.scatter(x = 'Ash', y = 'Label', title = 'Label vs Ash')
df.plot.scatter(x = 'Alcalinity of ash', y = 'Label',title = 'Label vs Alcalinity of ash')
df.plot.scatter(x = 'Magnesium', y = 'Label',title = 'Label vs Magnsium')
df.plot.scatter(x = 'Total phenols', y = 'Label',title = 'Label vs Total phenols' )
df.plot.scatter(x = 'Flavanoids', y = 'Label',title = 'Label vs Flavanoids')
df.plot.scatter(x = 'Nonflavanoid phenols', y = 'Label', title = 'Label vs Nonflavanoid phenols')
df.plot.scatter(x = 'Proanthocyanins', y = 'Label',title = 'Label vs  Proanthocyanins')
df.plot.scatter(x = 'Color intensity', y = 'Label',title = 'Label vs Color of intensity')
df.plot.scatter(x = 'Hue', y = 'Label', title = 'Label vs Hue')
df.plot.scatter(x = 'OD280/OD315 of diluted wines', y = 'Label')
df.plot.scatter(x = 'Proline', y = 'Label')


```

```python
data = df.values
Y = data[:,0]
X = data[:, 1:14]

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

ppn = Perceptron(max_iter = 1500, eta0 = 0.1, random_state = 1)
ppn.fit(X_train, Y_train)
y_pred_ppn = ppn.predict(X_test)
print("Accuracy score: " + str(accuracy_score(Y_test, y_pred_ppn)))
print("\nConfusion matrix: \n" + str(confusion_matrix(Y_test, y_pred_ppn)))
print("\nClassification report: \n" + str(classification_report(Y_test, y_pred_ppn)))
```

```python
from sklearn.linear_model import LogisticRegression
ppn = LogisticRegression(C = 100, max_iter = 1500, random_state = 1)
ppn.fit(X_train, Y_train)
y_pred_log_reg = ppn.predict(X_test)
print("Accuracy score: " + str(accuracy_score(Y_test, y_pred_log_reg)))
print("\nConfusion matrix: \n" + str(confusion_matrix(Y_test, y_pred_log_reg)))
print("\nClassification report: \n" + str(classification_report(Y_test, y_pred_log_reg)))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Plot confusion matrix for Perceptron
plt.figure(figsize=(6, 6))
cm_ppn = confusion_matrix(Y_test, y_pred_ppn)
sns.heatmap(cm_ppn, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Perceptron')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()

# Plot confusion matrix for Logistic Regression
plt.figure(figsize=(6, 6))
cm_log_reg = confusion_matrix(Y_test, y_pred_log_reg)
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python

import joblib


# Save the trained model
joblib.dump(ppn, 'perceptron_model.pkl')

# Load the saved model
loaded_model = joblib.load('perceptron_model.pkl')

# Predict classes for the new wine samples
new_samples = np.array([[13.72, 1.43, 2.5, 16.7, 108, 3.4, 3.67, 0.19, 2.04, 6.8, 0.89, 2.87, 1285],
                        [12.37, 0.94, 1.36, 10.6, 88, 1.98, 0.57, 0.28, 0.42, 1.95, 1.05, 1.82, 520]])
predicted_classes = loaded_model.predict(new_samples)

print("Predicted classes for the new wine samples:")
print(predicted_classes)
```
