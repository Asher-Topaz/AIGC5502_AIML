```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
file_name = 'operations.csv'
df = pd.read_csv(file_name)

df.head(50)
```

```python
#display df info
display(df.info())
display(df.describe())
display(df.shape)
display(df.dtypes)
```

```python
#print unique values of each column
cols = df.columns.tolist()
for col in cols:
    unique = df[col].unique()
    print('{col}: {unique}'.format(col = col, unique = unique))
```

```python
#check null percentage of each column and set a cutoff
#if column is greater than cutoff percentage delete it
cutoff_percentage = 85

row_count = len(df)
series_null = df.isnull().sum()
display(series_null)

cutoff_cols = []
for index, value in series_null.items():
    null_percentage = value/row_count * 100
    if(null_percentage>=cutoff_percentage):
        cutoff_cols.append(index)
    print('{ind}: {val}'.format(ind = index, val = null_percentage))
    
print("\n")
print('cutoff cols:', cutoff_cols)
```

```python
df.drop(columns=cutoff_cols, inplace=True)
print(df.shape)
```

```python
country_cols = df['Country'].unique()
country_cols
```

```python
#print list of unique planes belonging to each country
unique_planes_obj_list = {}
for country in country_cols:
    df_country = df[df['Country'] == country];
    unique_planes_obj_list[country] = df_country['Aircraft Series'].unique()

display(unique_planes_obj_list)
```

```python
#get a list of unique planes where country is null
df_country_null = df[df['Country'].isnull()]

unique_planes_null = df_country_null['Aircraft Series'].unique()
print(unique_planes_null)
```

```python
truly_unique_planes_obj = {}
for unique_plane in unique_planes_null:
    country_list = []
    for country in country_cols:
        if unique_plane in unique_planes_obj_list[country]:
            country_list.append(country)
    if len(country_list) == 1:
        truly_unique_planes_obj[unique_plane] = country_list[0]

print(truly_unique_planes_obj)
```

```python
#do the above for target country
unique_targets_obj_list = {}
for country in country_cols:
    df_country = df[df['Country'] == country];
    unique_targets_obj_list[country] = df_country['Target Country'].unique()

display(unique_targets_obj_list)
```

```python
unique_targets_null = df_country_null['Target Country'].unique()
print(unique_targets_null)
```

```python
#find truly unique target countries

truly_unique_targets_obj = {}
for unique_target in unique_targets_null:
    country_list = []
    for country in country_cols:
        if unique_target in unique_targets_obj_list[country]:
            country_list.append(country)
    if len(country_list) == 1:
        truly_unique_targets_obj[unique_target] = country_list[0]

print(truly_unique_targets_obj)
```

```python
#fill in the countries using unique target countries and planes
for plane in truly_unique_planes_obj:
    df.loc[df['Aircraft Series'] == plane, 'Country'] = truly_unique_planes_obj[plane]
```

```python
for target_country in truly_unique_targets_obj:
    df.loc[df['Target Country'] == target_country, 'Country'] = truly_unique_targets_obj[target_country]
```

```python
new_country_null_count = df['Country'].isnull().sum()
new_country_null_percentage = new_country_null_count/row_count * 100

print(new_country_null_count)
print(new_country_null_percentage)
```

```python
#country null percentage dropped from 29% to 18%
#dropping remaining null rows
df.dropna(subset=['Country'], inplace = True)

row_count = len(df)

print(df['Country'].isnull().sum())
print(row_count)

country_cols = df['Country'].unique()
```
